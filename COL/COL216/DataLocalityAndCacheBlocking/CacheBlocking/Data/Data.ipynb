```python
import pandas as pd
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
index = []
for i in range(0, 100):
    index.append(i*20 + 5)
len(index)
```

```python
data1 = pd.read_csv('seconds1.csv', index_col=False)
data2 = pd.read_csv('seconds2.csv', index_col=False)
data3 = pd.read_csv('seconds3.csv', index_col=False)
data4 = pd.read_csv('seconds4.csv', index_col=False)
data5 = pd.read_csv('seconds5.csv', index_col=False)
data = (data1 + data2 + data3 + data4 + data5) / 5
data.set_index(pd.Index(index), inplace=True)
data.dropna(inplace=True, axis=1)
data
```

```python
sns.lineplot(data=data, dashes=False, markers=False, palette="tab10", linewidth=2).set_xlabel('Matrix Size')
```
