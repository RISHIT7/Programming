```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_theme(style="darkgrid")
```

```python
index = []
for i in range(0, 51):
    index.append(i*20 + 10)
```

```python
data1 = (pd.read_csv('Gflops1.csv', index_col=False))
data2 = (pd.read_csv('Gflops2.csv', index_col=False))
data3 = (pd.read_csv('Gflops3.csv', index_col=False))
data4 = (pd.read_csv('Gflops4.csv', index_col=False))
```

```python
data = ((data1 + data2 + data3 + data4)/4)
data.set_index(pd.Index(index), inplace=True)
data.to_csv('Gflops.csv', index=False, float_format='%.3f')
```

```python
sns.lineplot(data=data, dashes=False, markers=False, palette="tab10", linewidth=2).set_xlabel('Matrix Size')
```
