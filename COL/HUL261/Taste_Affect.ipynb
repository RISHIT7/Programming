```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
DATA = pd.read_excel('../HUL261/Id_Affect.xlsx')
```

```python
DATA = pd.DataFrame(DATA)
DATA.head()
```

```python
DATA = DATA.drop(columns='Unnamed: 0')
```

```python
DATA.head()
```

```python
def print_foo(data_x, data_y, ax):
    y_ = DATA[data_x][1:]
    z_ = DATA[data_y][1:]
    print(z_.head())
    ax.set_label(data_y + ' vs ' + data_x)
    ax.set_xlabel(data_x)
    ax.set_ylabel(data_y)
    ax.scatter(y_, z_)
```

```python
fig, axes = plt.subplots(4, 2, figsize = (12, 16))

ax = axes[0][0]
print_foo('Fear', 'Anger', ax)

ax = axes[0][1]
print_foo('Fear', 'Sweet', ax)

ax = axes[1][1]
print_foo('Joy', 'Bitter', ax)

ax = axes[1][0]
print_foo('Sadness', 'Bitter', ax)

ax = axes[2][0]
print_foo('Bitter', 'Fear', ax)

ax = axes[2][1]
print_foo('Sour', 'Arousal (F)', ax)

ax = axes[3][0]
print_foo('Anger', 'Joy', ax)

ax = axes[3][1]
print_foo('Sweet', 'Valence (P)', ax)
```
