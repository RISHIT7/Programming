```python
import yahooquery as yq
```

```python
def get_PE(symbol):
    symbol += ".NS"
    stock_data = yq.Ticker(symbol)
    PE = stock_data.summary_detail[symbol]['trailingPE']
    return PE
```

```python
def get_MC(symbol):
    symbol += ".NS"
    stock_data = yq.Ticker(symbol)
    MC = stock_data.summary_detail[symbol]['marketCap']
    return MC
```
